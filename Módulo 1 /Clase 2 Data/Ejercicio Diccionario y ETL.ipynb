1) Abrir el archivo "Emisiones_CO2.csv" y cargar sus datos en un diccionario.

```python

archivo = open ('Emisiones_CO2.csv', 'r', encoding='latin-1')
```

```python
dicc_emisiones = {  'cod_pais' : [],
                    'nom_pais' : [],
                    'region' : [],
                    'anio' : [],
                    'co2' : [],
                    'co2_percapita' : []}
```

```python
for ingesta in archivo:
    
    lista = ingesta.split("|")
    dicc_emisiones["cod_pais"].append(lista[0])
    dicc_emisiones["nom_pais"].append(lista[1])
    dicc_emisiones["region"].append(lista[2])
    dicc_emisiones["anio"].append(lista[3])
    dicc_emisiones["co2"].append(lista[4])
    dicc_emisiones["co2_percapita"].append(lista[5])
    
```

```python
#¿Qué tipos de datos usar para cada una?
print(dicc_emisiones["cod_pais"][30]) # cualitativo - categórico
print(dicc_emisiones["nom_pais"][30]) # cualitativo - categórico
print(dicc_emisiones["region"][30]) # cualitativo - categórico
print(dicc_emisiones["anio"][30]) # cualitativo ordinal / cuantitativo discreto
print(dicc_emisiones["co2"][30]) # cuantitativo continuo
print(dicc_emisiones["co2_percapita"][30]) # cuantitativo continuo
```

```python
# ¿Qué tipo de variables de python son?

print(type(dicc_emisiones["cod_pais"][30]))
print(type(dicc_emisiones["nom_pais"][30]))
print(type(dicc_emisiones["region"][30]))
print(type(dicc_emisiones["anio"][30]))
print(type(dicc_emisiones["co2"][30]))
print(type(dicc_emisiones["co2_percapita"][30]))
```

```python
# ¿Hay valores faltantes?
nulo = 0
for indice, elemento in enumerate(dicc_emisiones['cod_pais']):
    if elemento == None:
        nulo += 1
print('La cantidad de nulos para cod pais es: ',nulo)
```

```python
nulo = 0
for indice, elemento in enumerate(dicc_emisiones['nom_pais']):
    if elemento == None:
        nulo += 1
print('La cantidad de nulos para nom pais es: ',nulo)
```

```python
nulo = 0
for indice, elemento in enumerate(dicc_emisiones['region']):
    if elemento == None:
        nulo += 1
print('La cantidad de nulos para region es: ',nulo)
```

```python
nulo = 0
for indice, elemento in enumerate(dicc_emisiones['anio']):
    if elemento == None:
        nulo += 1
print('La cantidad de nulos para anio es: ',nulo)
```

```python
## ETL - Limpieza de co2_percapita
for indice, elemento in enumerate(dicc_emisiones['co2_percapita']):
    # Convertir a string para reemplazar
    elemento = str(elemento).replace('.', '').replace(',', '.')
    
    # Convertir a float o None si está vacío
    if elemento.strip():
        dicc_emisiones['co2_percapita'][indice] = float(elemento)
    else:
        dicc_emisiones['co2_percapita'][indice] = None

## Conteo de valores nulos
nulo = dicc_emisiones['co2_percapita'].count(None)
print('La cantidad de nulos para co2_percapita es:', nulo)
```

```python
for indice, elemento in enumerate(dicc_emisiones['co2']):
    elemento = str(elemento).replace('.', '').replace(',', '.')
    if (elemento != ''):
        elemento = float(elemento)
    else:
        elemento = None
    dicc_emisiones['co2'][indice] = elemento
```

```python
nulo = 0
for indice, elemento in enumerate(dicc_emisiones['co2']):
    if elemento == None:
        nulo += 1
print('La cantidad de nulos para co2 es: ',nulo)
```

```python
#¿Cuál es el total de emisiones de CO2 para 'América Latina y Caribe' en el año 2010?

import enum


filtro_region = 'América Latina y Caribe'
filtro_anio = '2010'
emisiones = 0
for indice, elemento in enumerate(dicc_emisiones['region']):
    if (dicc_emisiones['region'][indice] == filtro_region):
        if (dicc_emisiones['anio'][indice] == filtro_anio):
            if (dicc_emisiones['co2'][indice] != None):
                emisiones += dicc_emisiones['co2'][indice]
print('las emisiones en: ', filtro_region, 'en ', filtro_anio, 'fueron de ',emisiones, 'kt')
```
