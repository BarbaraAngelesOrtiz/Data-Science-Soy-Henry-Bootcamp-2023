## Graficaremos la función $$y=mx+b,$$ donde $m$ es la pendiente y $b$ es el intercepto en el origen

```python
x = np.linspace(0,2,30)
```

```python
print(x)
```

```python
m = 2 # pendiente
b = -4 # intercepto
y = m*x + b
```

## Para graficar llamamos al método plot, que en español significa graficar, en este caso inicialmente necesita de dos parámetros, el primero los datos de la coordenada $X$ y como segundo parámetro la coordenada $Y$

## en este caso se necesita ejecutar en pantalla la gráfica, para eso, hacemos uso del metodo $show()$.
```python
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.show()
```

# Agregamos un nombre en la coordenada X de la siguente manera:

```python
plt.plot(x, y)
plt.xlabel("Coordenada X") # con esta consigna agregamos a la coordenada x un nombre
plt.show()
```

## Luego también a la coordenada $Y$ le colocamos un nombre con ylabel("")

```python
plt.plot(x, y)
plt.xlabel("Coordenada X")
plt.ylabel("Esta es la coordenada Y") # con esta consigna agregamos a la coordenada y un nombre
plt.show()
```

## También podemos colocarle un nombre a la gŕafica de la siguente forma con $plt.title()$

```python
plt.plot(x, y)
plt.xlabel("Coordenada X")
plt.ylabel("Esta es la coordenada Y") 
plt.title("Este es el nombre de la gŕafica") # De esta forma colocamos un nombre a la gráfica
plt.show()
```

## En ciertos casos para una mejor lectura de la gráfica, es necesario colocar grillas (cuadros por unidades a la imagen). Esto es posible mendiante el método $plt.grid()$

```python
plt.grid()
```

## finalmente, si deseamos cambiarle el color a la gráfica, en este caso, adicionamos parámetros a $plt.plot()$ de la siguente forma:

```python
plt.plot(x, y, "r") # ponemos que el color de la gráfica sea de color rojo en el tercer parámetro
plt.xlabel("Coordenada X")
plt.ylabel("Esta es la coordenada Y") 
plt.title("Este es el nombre de la gŕafica")
plt.grid() 
plt.show()
```

## Existen múltiples colores para agregar a la grárica en este caso estos son algunos:
1. "r" rojo (red)
2. "g" verde (green)
3. "y" amarillo (yellow)
4. "k" negro (key)

## Si deseamos agregar dos o mas gráficas en un misma imagen hacemos lo siguente:

## vamos a graficar una tendencia cuadrática de la siguiente forma: $$y_{1}= a*x^2 + b$$

```python
y1 =  3*(x**2) + 1
```

## y generamos la tendencia lineal y la tendencia cuadrática en la siguiente gráfica de la siguiente forma:

```python
plt.plot(x, y) # tendencia lineal
plt.plot(x, y1) # tendencia cuadrática
plt.grid()
plt.show()
```

## de la anterior gráfica notamos que la curva amarilla pertenece a la tendencia cuadrática y la linea pertenece a la tendencia lineal.

## En este caso que no queramos que nos grafique una curva o una linea si no simplemente los puntos de los datos a graficar usamos el siguiente metodo $plt.scatter()$

```python
plt.plot(x, y) 
plt.scatter(x, y1) # simplemente en vez de plot usamos scatter. 
plt.grid()
plt.show()
```

## si deseamos cambiarle el color en scatter, usamos el parámetro color de la siguiente forma

```python
plt.plot(x, y) 
plt.scatter(x, y1, color = "k") # de esta manera le colocamos color negro (key)
plt.grid()
plt.show()
```

## es importante que podemos generar 2 o más curvas o puntos dentro de la gráfica, por ejemplo

```python
plt.plot(x, y) 
plt.scatter(x, y1, color = "k")
plt.plot(y1, x, "cyan") #nueva curva
plt.scatter(y, x, color = "y") #nuevos puntos
plt.grid()
plt.show()
```

## aprovechando la gráfica anterior, supongamos que solo deseamos graficar en la coordenada x desde -2.5 a 5.0 e ignorar la región sobrante, para eso, hacemos uso de plt.xlim() de la siguiente forma:

```python
plt.plot(x, y) 
plt.scatter(x, y1, color = "k")
plt.plot(y1, x, "cyan") 
plt.scatter(y, x, color = "y") 
plt.grid()
plt.xlim(-2.5, 5.0) # el primer parámetro hace refencia al límite izquierdo y el segundo parámetro al límite derecho
plt.show()
```

## también podemos aplicar dicho límite en la coordenada y, mediante $ylim$

```python
plt.plot(x, y) 
plt.scatter(x, y1, color = "k")
plt.plot(y1, x, "cyan") 
plt.scatter(y, x, color = "y") 
plt.grid()
plt.ylim(0, 10) # el primer parámetro hace refencia al límite inferior y el segundo parámetro al límite superior
plt.show()
```

## también combinar los dos limites a la vez

```python
plt.plot(x, y) 
plt.scatter(x, y1, color = "k")
plt.plot(y1, x, "cyan") 
plt.scatter(y, x, color = "y") 
plt.grid()
plt.xlim(-1, 2) # limite en la coordenada x
plt.ylim(0, 10) # limite en la coordenada y
plt.show()
```

# Histogramas

## Uno de los pilares importantes a la hora de analizar fenómenos mediante de la estadistica son los histogramas.

## Los histogramas facilitan la vida a los estadistas a la hora de ver gráficamente comportamiento de datos y poder obtener una conclusión posible de el porqué se comportan así.

## En palabras simples del autor de este notebook, un histograma es una gráfica donde se observa la cantidad de veces en la cuales se repite un número. En este caso se coloca en la coordenada x el numero que quiero saber cuantas veces se repite y en la coordenada y las veces que se repite. En este caso no puede ser únicamente ver cuántas veces se repite un número entero o flotante, también puede verse cuántas veces se repite una región de números especifica, cuantas veces se repite un nombre o una palabra. Veremos como se construyen estas gráficas en matplotlib.

## inicialmente construimos una lista la cual se repitan un conjunto de números del 1 al 10

```python
lista = [2, 2, 2, 5, 5, 6, 7, 10, 10, 10, 10]
```

## usamos plt.hist() para hacer el histograma de la lista anterior de la siguiente forma.

```python
plt.hist(lista) # ingresamos la lista dentro
plt.show()
```

## como notamos en la gráfica, el número 2 se repite 3 veces, el número 8 se repite 0 veces, el 10 se repite 4 veces

## también muestra en este caso elementos como palabras. por ejemplo con la siguiente lista construirémos un histograma, en este caso con nombres que se reppiten

```python
lista =  ["Pablo", "Pablo", "Pablo", "Andrea", "Andrea", "Harold"]
```

## usamos el mismo método de matplotlib para obtener el histograma

```python
plt.hist(lista)
plt.plot()
```

## También podemos aplicar titulos, grillas colores como en el caso anterior de las curvas y lineas, también graficar multiples histogramas. Para este caso final, usaremos dos listas.

```python
lista1 = [2, 2, 2, 5, 5, 6, 7, 10, 10, 10, 10]
lista2 = [-2, -2, -2, -5, -5, -6, -7, -10, -10, -10, -10]
```

```python
plt.hist(lista1, color = "r") # primer histograma rojo
plt.hist(lista2, color = "y") # segundo histograma amarillo
plt.grid() # aplimamos grilla
plt.xlabel("Números") # aplicamos un titulo a la coordenada x
plt.ylabel("Conteo") # aplicamos un titulo a la coordenada y
plt.title("Superhipermega histograma cool") # un título a la gráfica
plt.show()
```
